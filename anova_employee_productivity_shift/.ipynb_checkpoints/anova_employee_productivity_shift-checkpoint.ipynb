```python
import numpy as np
import pandas as pd
morning = np.random.normal(loc=52, scale=3, size=30)
afternoon = np.random.normal(loc=50, scale=2, size=30)
night = np.random.normal(loc=47, scale=4, size=30)
df=pd.DataFrame({'Employee ID':range(1,31),'Morning Tasks':morning,'Afternoon Tasks':afternoon,'Night Tasks':night})
df['Morning Tasks']=df['Morning Tasks'].apply(lambda x:round(x,0))
df['Afternoon Tasks']=df['Afternoon Tasks'].apply(lambda x:round(x,0))
df['Night Tasks']=df['Night Tasks'].apply(lambda x:round(x,0))
df
```

```python
from scipy.stats import f_oneway
f_stat,p_value=f_oneway(df['Morning Tasks'],df['Afternoon Tasks'],df['Night Tasks'])
p_value
```

so at least one of the categories is significantly different from others.

```python
df_without_ID=df.drop(columns='Employee ID')
```

```python
df_melted=df_without_ID.melt(var_name='Shift',value_name='No. of Tasks')
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
tukey=pairwise_tukeyhsd(endog=df_melted['No. of Tasks'],groups=df_melted['Shift'],alpha=0.05)
print(tukey)
```

```python
# so from the tukey hsd I conclude the productivity is max in the morning and it's statistically 
# significant not due to randomness.
```
